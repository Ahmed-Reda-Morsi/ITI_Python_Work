# 04_Lab: dictionary
 

```python
#take sensors reads sperarated with " "
s1,s2,s3= input("enter three sensors values as s1 s2 s3").split(" ",3)

# cast values to float
s1=float(s1)
s2=float(s2)
s3=float(s3)

# define dictionary and append values  
Sensor={}
Sensor["s1"]=s1
Sensor["s2"]=s2
Sensor["s3"]=s3
#define list with values
sensor_list=[s1,s2,s3]
#define tuples with values
sensor_tuple=(s1,s2,s3)

#print all 
print("dictionary:  ",Sensor)
print("List:        ",sensor_list)
print("tuple:       ",sensor_tuple)
```
